## 1. 初始化客户端
配置阿里云百炼（DashScope）的API Key，并初始化OpenAI兼容客户端。

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# 加载环境变量
load_dotenv()

# 创建OpenAI客户端，指向阿里云百炼的兼容接口
client = OpenAI(
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1"
)
```

## 2. 调用大模型
向模型发送对话请求。这里使用 `qwen-plus` 模型（也可替换为其他百炼支持的模型，如 `deepseek-v3` 等）。

```python
print("🚀 正在调用大模型...")
response = client.chat.completions.create(
    model="qwen-plus",
    messages=[
        {"role": "system", "content": "你是一名友好的AI助教。"},
        {"role": "user", "content": "当前时间是多少"}
    ],
    stream=False
)
```

## 3. 打印响应内容
提取并打印模型返回的具体文本，解决之前 response 未定义的问题。

```python
# 打印完整的响应对象
print("【完整响应】:")
print(response)

print("\n" + "-" * 50 + "\n")

# 提取并打印模型回复的文本内容
print("【模型回复】:")
print(response.choices[0].message.content)
```
